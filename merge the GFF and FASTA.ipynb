# Step 1: Import Necessary Libraries

```python
import os
import glob
import shutil
```

# Step 2: Set Up Paths
Define the path where your files are located and collect all the relevant files.

```python
folder_path = 'data/extracted_genomes/'
gff3_output_file = os.path.join(folder_path, 'merged.gff3')
fasta_output_file = os.path.join(folder_path, 'merged.fasta')
```

# Step 3: Create a Function to Merge Files
This function will take pairs of GFF and FASTA files, merge their contents, and store the output in a new file.

```python
def merge_files(file_list, output_file):
    with open(output_file, 'w') as outfile:
        for fname in file_list:
            with open(fname) as infile:
                outfile.write(infile.read() + '\n')  # Write content to output file
    print(f'Merged {len(file_list)} files into {output_file}')
```

# Step 4: Collect GFF3 and FASTA Files
Use glob to find all GFF3 and FASTA files in the specified directory.

```python
gff3_files = glob.glob(os.path.join(folder_path, '*.gff3'))
fasta_files = glob.glob(os.path.join(folder_path, '*.fasta'))
```

# Step 5: Merge GFF3 and FASTA Files
Call the merge function for both GFF3 and FASTA files.

```python
# Merge GFF3 files
merge_files(gff3_files, gff3_output_file)

# Merge FASTA files
merge_files(fasta_files, fasta_output_file)
```

# Step 6: Verify the Merging
After merging, you can check the sizes of the merged files against the unmerged files to ensure correctness. Here’s how you can do that:

```python
def verify_merge(original_files, merged_file):
    original_size = sum(os.path.getsize(f) for f in original_files)
    merged_size = os.path.getsize(merged_file)

    if original_size == merged_size:
        print(f'The merged file {merged_file} matches the total size of original files: {original_size} bytes.')
    else:
        print(f'Size mismatch! Merged size: {merged_size} bytes. Expected: {original_size} bytes.')

# Verify for GFF3 files
verify_merge(gff3_files, gff3_output_file)

# Verify for FASTA files
verify_merge(fasta_files, fasta_output_file)
```
