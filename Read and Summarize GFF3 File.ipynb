```python
import pandas as pd

def parse_gff3(file_path):
    """
    Parses a GFF3 file to summarize feature types and attributes.
    Args:
    - file_path: Path to the GFF3 file.

    Returns:
    - features_summary: A DataFrame summarizing the feature types.
    - attributes_summary: A dictionary of unique attribute keys.
    """
    feature_types = []
    attributes_list = []

    with open(file_path, "r") as file:
        for line in file:
            if line.startswith("#"):
                continue  # Skip comments and headers
            parts = line.strip().split("\t")
            if len(parts) != 9:
                continue  # Skip malformed lines
            feature_types.append(parts[2])  # Feature type (e.g., gene, exon)
            attributes_list.append(parts[8])  # Attributes field

    # Summarize feature types
    feature_types_df = pd.DataFrame(feature_types, columns=["Feature Type"])
    features_summary = feature_types_df.value_counts().reset_index()
    features_summary.columns = ["Feature Type", "Count"]

    # Extract unique attribute keys
    attributes_summary = set()
    for attr in attributes_list:
        attr_pairs = attr.split(";")
        for pair in attr_pairs:
            key_value = pair.split("=")
            if len(key_value) == 2:
                attributes_summary.add(key_value[0])

    return features_summary, attributes_summary


# Example usage
gff3_file = "data/extracted_genomes/merged.gff3"  # Replace with your GFF3 file path
features_summary, attributes_summary = parse_gff3(gff3_file)

# Print summary of feature types
print("Feature Types Summary:")
print(features_summary)

# Print attribute keys
print("\nUnique Attribute Keys in GFF3:")
print(attributes_summary)

# Optionally save output
features_summary.to_csv("gff3_features_summary.csv", index=False)
print("\nFeature types saved to 'gff3_features_summary.csv'.")
```
